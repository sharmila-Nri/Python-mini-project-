```python
ScoredAboveZero=df['Score']>'0-0'
```

```python
df[ScoredAboveZero]
```

```python
ScoredByBrazil=df['home_team']=='Brazil'
```

```python
df[ScoredAboveZero & ScoredByBrazil]
```

```python
Brazildf=df[ScoredAboveZero & ScoredByBrazil].sort_values(by='home_score', ascending=True)
```

```python
Brazildf
```

```python
Brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
Brazildf.plot(x='home_score',y='away_score',kind='scatter')
plt.xlabel('Home_score')
plt.ylabel('away_score')
plt.legend('Takeaway score')
plt.title('Brazil Matches ')
```

```python
Brazildf.plot(x='Date',y='home_score',kind='bar')
plt.xlabel('Date')
plt.ylabel('home_score')
plt.legend('Takeaway score')
plt.title('Brazil Matches ')
```
